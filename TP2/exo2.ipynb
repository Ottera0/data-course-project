```python
from PIL import Image
import matplotlib.pyplot as plot

imgfile = Image.open("images/flower.jpg")
histogram = imgfile.histogram()
red = histogram[0:255]
green = histogram[256:511]
blue = histogram[512:767]
x = range(255)
y = []
for i in x:
    y.append((red[i], green[i], blue[i]))
figure, axes = plot.subplots()
axes.set_prop_cycle("color", ["red", "green", "blue"])
plot.plot(x, y)
plot.show()
```

Votre prochaine question consiste  obtenir les 20 premires intensits dans chaque bande et  crer un seul trac de ces premires intensits. crivez un programme python qui peut raliser cela.

```python
from PIL import Image
import matplotlib.pyplot as plot

# Ouvrir l'image
imgfile = Image.open("images/flower.jpg")

# Obtenir l'histogramme de l'image
histogram = imgfile.histogram()

# Extraire les 20 premires intensits pour chaque bande
red_first20 = histogram[0:20]
green_first20 = histogram[256:276]
blue_first20 = histogram[512:532]

# Dfinir l'axe des x pour les 20 premires intensits
x = range(20)

# Crer une figure et des axes pour le trac
fig, ax = plot.subplots()

# Tracer les 20 premires intensits pour chaque bande
ax.plot(x, red_first20, color="red", label="Rouge")
ax.plot(x, green_first20, color="green", label="Vert")
ax.plot(x, blue_first20, color="blue", label="Bleu")

# Ajouter une lgende pour distinguer les bandes
ax.legend()

# Afficher le graphique
plot.show()
```
