```python
from pandas import json_normalize
import pandas as pd
import json
import matplotlib.pyplot as plot

data = json.load(open("data/pl.json"))
dataframe = json_normalize(data)
grouped = dataframe.groupby("year").count()
grouped = grouped.rename(columns={"languageLabel": "count"}).reset_index()
grouped.plot(x=0, kind="bar", title="Programming languages per year")
```

```python
from pandas import json_normalize
import pandas as pd
import json
import matplotlib.pyplot as plt

# Chargement des donnes JSON
with open("data/plparadigm.json") as file:
    jsondata = json.load(file)

# Transformation des donnes en DataFrame
dataframe = pd.DataFrame(jsondata)

# Conversion du type de donnes de la colonne 'year'
dataframe['year'] = pd.to_numeric(dataframe['year'])

# Agrgation des donnes par anne et paradigme, comptant le nombre de langages pour chaque groupe
grouped = dataframe.groupby(['year', 'paradigmLabel']).size().unstack(fill_value=0)

# Trac des donnes agrges
fig, ax = plt.subplots(figsize=(14, 8))
grouped.plot(kind='bar', stacked=True, ax=ax)
plt.title('Nombre de langages de programmation par paradigme et par anne')
plt.xlabel('Anne')
plt.ylabel('Nombre de langages')
plt.legend(title='Paradigme')
plt.show()
```
